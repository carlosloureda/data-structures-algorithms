````python
daysOfMonth = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

def isLeapYear(year):
    # Año bisiesto es el divisible entre 4, salvo que sea año secular 
    #-último de cada siglo, terminado en «00»-, en cuyo caso también ha de ser divisible entre 400.
    return year % 400 == 0 if year % 100 == 0 else year % 4 ==0 
    

def _test_isLeapYear():
    assert isLeapYear(8) == True
    assert isLeapYear(9) == False
    assert isLeapYear(400) == True    
    assert isLeapYear(600) == False
    assert isLeapYear(2019) == False
    assert isLeapYear(2004) == True
    
_test_isLeapYear() 

def daysBetweenDates (y1, m1, d1, y2, m2, d2):
    m1 -= 1
    m2 -= 1
    days = 0
    daysOfMonthY1 = daysOfMonth.copy()
    daysOfMonthY2 = daysOfMonth.copy()
    # Case 1: dates are on different year
    if y1 != y2:           
        if isLeapYear(y1):       
            daysOfMonthY1[1] = 29
        if isLeapYear(y2):
            daysOfMonthY2[1] = 29
        # 1. calculate days until end of year y1
            #1.1 from next month to December
        for month in range(m1+1, 12):
            #print("month: ", month, " daysOfMonthY1[month]: ", daysOfMonthY1[month])
            days += daysOfMonthY1[month]
            if month == 12: 
                days -= 1 # don't include last day of year
            #1.2 from d1 to last day of that month
        days += daysOfMonthY1[m1] - d1
        
        # 2. calculate days y2 until day and month
        for month in range(0, m2):
            days += daysOfMonthY2[month]
            #1.2 days in that month
        days += d2
        # 3. calculate the days in the years in between
        for year in range(y1+1, y2):
            days += 366 if isLeapYear(year) else 365
    # Case 2: dates are on different month (same year)
    elif m1 != m2:        
        if isLeapYear(y2):
            daysOfMonthY2[1] = 29
        # 1. calculate days d1 to end of month
        days = daysOfMonthY2[m1] - d1 # + 1
        # 2. calculate days from beginning of month to d2
        days += d2
        # 3. Calculate the days in the months (cargefull on february)
        for month in range(m1+1, m2):
            days += daysOfMonthY2[month]           
        
    # Case 3: Dates are on same year and month (just return the difference)
    else:
        days = d2 - d1
    return days


def _test_DaysBetweenDates():
    assert(daysBetweenDates(2012,1,1,2012,2,28) == 58)
    assert(daysBetweenDates(2011,6,30,2012,6,30) == 366) # 367 calculando dia final
    assert(daysBetweenDates(2011,1,1,2012,8,8) == 585) # 586 (incluido el dia final)
    assert(daysBetweenDates(1900,1,1,1999,12,31) == 36523) # 36524 (incluido el dia final) me da 36525
            
def test():
    _test_isLeapYear()
    _test_DaysBetweenDates()
    print("All test passed")
test()

````

````python
###
### Define a simple nextDay procedure, that assumes
### every month has 30 days.
###
### For example:
###    nextDay(1999, 12, 30) => (2000, 1, 1)
###    nextDay(2013, 1, 30) => (2013, 2, 1)
###    nextDay(2012, 12, 30) => (2013, 1, 1)  (even though December really has 31 days)
###

def nextDay(year, month, day):
    """
    Returns the year, month, day of the next day.
    Simple version: assume every month has 30 days.
    """
    _next_day = (year, month, day+1)
    if day == 30:
        _next_day = (year, month+1, 1) if month < 12 else (year+1, 1, 1)        
    return _next_day

print ( nextDay(2012, 12, 30) )
````
